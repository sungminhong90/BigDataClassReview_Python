```python
import pandas as pd
```

#### 엑셀파일 데이터 읽기

```python
xls_file = './data/test_score.xlsx'
```

```python
#이거는 엑셀 첫번째 시트 가져온다
df = pd.read_excel(xls_file)
df
```

```python
#sheet_name =1은 엑셀 2번째 시트를 가져온다
#(파이선 인덱스는 1부터 실행)
df = pd.read_excel(xls_file, sheet_name =1)
df
```

```python
df= pd.read_excel(xls_file, sheet_name = '기말고사')
df
```

```python
#index_col = 0 => '학생' 열을 인덱스 0으로 가져온다
df = pd.read_excel(xls_file, sheet_name = '기말고사', index_col = 0)
df
```

```python
#위와 같은 형식으로 중간고사 가져오기
df = pd.read_excel(xls_file, sheet_name = '중간고사', 
                  index_col = '학생')
df
```

### 데이터를 엑셀파일 쓰기
- 성적데이터1

```python
excel_exam_data1 = {
    '학생': ['A', 'B', 'C', 'D', 'E', 'F'],
    '국어': [80, 90, 95, 70, 75, 85],
    '영어': [90, 95, 70, 85, 90, 95],
    '수학': [85, 95, 75, 80, 85, 100]
}

df1 = pd.DataFrame(excel_exam_data1, columns = ['학생', '국어', '영어', '수학'])
df1
#columns 설정 이유는 순서를 저대로 하기 위해서-
```

```python
excel_writer = pd.ExcelWriter('./data/test_score2.xlsx',
                             engine = 'xlsxwriter')
df1.to_excel(excel_writer, index = False)
excel_writer.save()
```

```python
excel_writer = pd.ExcelWriter('./data/test_score_2_2.xlsx',
                             engine = 'xlsxwriter')
df1.to_excel(excel_writer, index = True)
excel_writer.save()
#이렇게 쓰면 index 0,1,2,3,4도 엑셀시트로 따라온다
```

```python
excel_writer = pd.ExcelWriter('./data/test_score3_xlsx', engine = 'xlsxwriter')
df1.to_excel(excel_writer, index = False, sheet_name='중간고사')
excel_writer.save()
##sheetname이 설정된다
```

- 성적데이터2

```python
excel_exam_data2 = {
    '학생': ['A', 'B', 'C', 'D', 'E', 'F'],
    '국어': [85, 95, 75, 80, 85, 100],
    '영어': [80, 90, 95, 70, 75, 85],
    '수학': [90, 95, 70, 85, 90, 95]
}

df2 = pd.DataFrame(excel_exam_data2,columns=['학생','국어','영어','수학'] )
df2
```

```python
#2개의 시트로 굽기 (midterm/final)
excel_writer = pd.ExcelWriter('./data/test_score4.xlsx',
                             engine = 'xlsxwriter')
df1.to_excel(excel_writer, index = False, sheet_name = '중간고사')
df2.to_excel(excel_writer, index = False, sheet_name = '기말고사')
excel_writer.save()
```

```python
##위의 엑셀을 index 포함해서 구어보기 -text_score4_4로 저장
excel_writer=pd.ExcelWriter('./data/test_score4_4.xlsx',
                           engine = 'xlsxwriter')
df1.to_excel(excel_writer, index =True, sheet_name = '중간고사')
df2.to_excel(excel_writer, index = True, sheet_name = '기말고사')
excel_writer.save()
```
