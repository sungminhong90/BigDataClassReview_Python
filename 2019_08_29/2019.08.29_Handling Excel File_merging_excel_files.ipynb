#### 여러 개의 엑셀 파일에서 데이터 수정

```python
file_name = './data/담당자별_판매실적_윤종필사원.xlsx'

new_file_name = re.sub('.xlsx', '2.xlsx', file_name)
new_file_name
```

```python
# 원하는 문자열이 포함된 파일을 검색해 리스트를 할당
excel_data_files = glob.glob("data/담당자별_판매실적_*.xlsx")
excel_data_files
```

```python
# 리스트에 있는 엑셀 파일만큼 반복 수행한다.

for f in excel_data_files:
    #엑셀 파일에서 dataframe의 형식으로 가져온다.
    df = pd.read_excel(f)
    
    #특정 열의 값을 변경한다.
    if(df.loc[1,'담당자']=='윤종필'):
        #loc[0-4] does not matter since it is in the same format : without position 
        # means to change all in the '담당자' column 
        df['담당자'] ='윤종필사원'
    elif(df.loc[1,'담당자']=='홍성민'):
        df['담당자']='홍성민대리'
    elif(df.loc[1, '담당자']=='신은영'):
        df['담당자']='신은영과장' 
    elif(df.loc[1, '담당자']=='김재경'):
        df['담당자']='김재경사원' 
    
    # 엑셀 파일 이름에서 지정된 문자열 패턴을 찾아서 파일명을 변경한다.
    f_new = re.sub('.xlsx', '_v2.xlsx', f)
    print(f_new)
    
     # 수정된 데이터를 새로운 이름의 엑셀 파일로 저장한다.
    new_excel_file = pd.ExcelWriter(f_new, engine='xlsxwriter')
    df.to_excel(new_excel_file, index = False)
    new_excel_file.save()
```

```python
df= pd.read_excel('./data/마케팅팀_판매실적_통합.xlsx')

#제품명 열에서 갤럭시S10이 있는 행만을 선택
product_name = '갤럭시S10'
galaxy=df[(df['제품명']==product_name)]

# 행별로 합계를 구하고 마지막 열 다음에 추가한다.
galaxy_sum = pd.DataFrame(galaxy.sum(axis=1), columns = ['연간판매량'])
galaxy_total = galaxy.join(galaxy_sum)

# 열별로 합해 분기별 합계와 연간판매량 합계를 구하고 마지막 행 다음에 추가한다.
galaxy_sum2 =pd.DataFrame(galaxy_total.sum(), columns = ['합계'])
galaxy_total2 = galaxy_total.append(galaxy_sum2.T)

# 지정된 항목의 문자열을 변경한다.
galaxy_total2.loc['합계', '제품명'] = product_name
galaxy_total2.loc['합계', '담당자'] = '전체'
galaxy_total2.loc['합계', '지역'  ] = '전체'

# 결과를 확인한다.
galaxy_total2
```

```python
def get_product_total(xls_file, product_name):
    # 제품명 열에서 갤럭시S10이 있는 행만 선택
    # product_name = '갤럭시S10'
    galaxy= df[(df['제품명']== product_name)]

    # 행별로 합계를 구하고 마지막 열 다음에 추가한다.
    galaxy_sum = pd.DataFrame(galaxy.sum(axis=1), columns = ['연간판매량'])
    galaxy_total = galaxy.join(galaxy_sum)

    # 열별로 합해 분기별 합계와 연간판매량 합계를 구하고 마지막 행 다음에 추가한다.
    galaxy_sum2 = pd.DataFrame(galaxy_total.sum(), columns=['합계'])
    galaxy_total2  = galaxy_total.append(galaxy_sum2.T)

    # 지정된 항목의 문자열을 변경한다.
    galaxy_total2.loc['합계', '제품명'] = product_name
    galaxy_total2.loc['합계', '담당자'] = '전체'
    galaxy_total2.loc['합계', '지역'  ] = '전체'

    # 결과를 확인한다.

    return galaxy_total2
```

```python
xls_file = './data/마케팅팀_판매실적_통합.xlsx'
prd_name = '갤럭시S10'

galaxy_total = get_product_total(xls_file, prd_name)
galaxy_total
```
